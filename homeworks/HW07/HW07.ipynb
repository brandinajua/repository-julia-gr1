# HW07 — Кластеризация

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
```

## Загрузка данных

```

df = pd.read_csv("S07-hw-dataset-01.csv")
df.head()
```

## Масштабирование

```

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

## KMeans

```

kmeans = KMeans(n_clusters=3, random_state=42)
labels_kmeans = kmeans.fit_predict(X_scaled)

sil_km = silhouette_score(X_scaled, labels_kmeans)
db_km = davies_bouldin_score(X_scaled, labels_kmeans)

sil_km, db_km
```

## DBSCAN

```

dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_db = dbscan.fit_predict(X_scaled)

mask = labels_db != -1

sil_db = silhouette_score(X_scaled[mask], labels_db[mask])
db_db = davies_bouldin_score(X_scaled[mask], labels_db[mask])

sil_db, db_db
```

## PCA и визуализация

```

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_kmeans)
plt.title("KMeans clusters (PCA)")
plt.show()

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_db)
plt.title("DBSCAN clusters (PCA)")
plt.show()
```
